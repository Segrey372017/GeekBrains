```
import numpy as np
from math import factorial as fact
from math import exp
```

```
def combinations(n, k):
    """Вычисляет сочетания k элементов из множества n"""
    return int(fact(n)/(fact(k)*fact(n-k)))
```

```
def p_binomial(k,n,p):
    """ Вычисляет вероятность соответствующего значения случайной величины из отрезка от 0 до n. Формула Бернулли"""
    return combinations(n,k)*p**k*(1-p)**(n-k)
```

```
def p_puassons(n, m,p):
    """ Вычисляет вероятность значений случайной величины (что событие произойдет m раз в n  испытаниях),
     при количестве испытаний n, вероятности события p. Для распределения Пуассона """
    lambda_ = n*p # интенсивность случайной величины
    return (lambda_**m/fact(m))*exp(-lambda_)
```

<h3> Задача1</h3>
<p> Вероятность того, что стрелок попадет в мишень, выстрелив один раз, равна 0.8. Стрелок выстрелил 100 раз. Найдите вероятность того, что стрелок попадет в цель ровно 85 раз. </p>

```
p85 = p_binomial(85, #случайная величина приняла значение 85 в 100 испытания
                100, # произошло 100 испытаний
                0.8) # вероятность наступления события - попадания в мишень)
print("Ответ: Вероятность того что стрелок попадет 85 раз в мишень, сделав 100 выстрелов составляет ", p85)    
```

<h3> Задача2 </h3>
<p> Вероятность того, что лампочка перегорит в течение первого дня эксплуатации, равна 0.0004. В жилом комплексе после ремонта в один день включили 5000 новых лампочек. <ul> <li> Какова вероятность, что ни одна из них не перегорит в первый день? </li> <li> Какова вероятность, что перегорят ровно две? </li> </ul> </p>

```
# Вычислим вероятность что в первый день не перегорит не одной лампочки
m0 = 0 # событие произойдет 0 раз (лампочка не перегорит) в течении первого дня
n2 = 5000 # количество испытаний (лампочек)
p2 = 0.0004 # вероятность наступления события (лампочка перегорела)

p0 = p_puassons(n2, m0, p2)
print(f"Вероятность, что в течение первого дня эксплуатации не перегорит ни одной лампочки составляет {p0}")
```

```
#Вычислим вероятность, что перегорит ровно 2 лампочки
m2 = 2 # Перегорит 2 лампочки в течение дня

p2 = p_puassons(n2, m2, p2)
print(f"Вероятность, что в течение первого дня эксплуатации перегорят ровно 2 лампочки составляет {p2}")
```

<h3> Задача3 </h3>
<p> Монету подбросили 144 раза. Какова вероятность, что орел выпадет ровно 70 раз?</p>

```
# Вычислим вероятность, применив формулу Бернулли
p3 = p_binomial(70, # вычисляем, что случайная величина примет значение 70 при подбрасываннии монеты 144 раза
                144, # количество подбрасываний в одном опыте было 144 раза
                0.5) # вероятность, что при броске монеты выпадет орел 1/2
print("Ответ: вероятность что при 144 бросках, орел выпадет 70 раз состаляет ", p3)
```

<h3> Задача4</h3>
<p> В первом ящике находится 10 мячей, из которых 7 - белые. Во втором ящике - 11 мячей, из которых 9 белых. Из каждого ящика вытаскивают случайным образом по два мяча. <ul>  <li>Какова вероятность того, что все мячи белые? </li> <li>Какова вероятность того, что ровно два мяча белые? </li> <li>Какова вероятность того, что хотя бы один мяч белый?</li> </ul></p>

```
# Определим вероятность что все выбранные мячи белые
box1 = {'w': 7, 'b': 3}
box2 = {'w': 9, 'b': 2}
bs1 = sum(box1.values())
bs2 = sum(box2.values())
print(bs1, '\t', bs2)
```

```
# Вытащить 4 белых мяча значит, что из каждого ящика необходимо вытащить 2 белых мяча

mA1 = combinations(box1['w'], 2) # сочетания выбрать 2 мяча из 7 имеющихся в первом ящике - благоприятные исходы
nA1 = combinations(bs1, 2) # все возможные исходы - вытащить 2 мяча
pA1 = mA1/nA1

#Аналогичный расчет для второго ящика
mA2 = combinations(box2['w'], 2) # сочетания выбрать 2 мяча из 9 имеющихся в  ящике - благоприятные исходы
nA2 = combinations(bs2, 2) # все возможные исходы - вытащить 2 мяча
pA2 = mA2/nA2

#Поскольку необходимо извлеч 2 белых мяча из обоих ящиков то вероятности перемножаются
pAB1 = pA1*pA2
print("Вероятность того, что все мячи белые состаляет ", pAB1)
```

```
#Вероятность того, что ровно 2 мяча белые

# Здесь имеется 2 события 1 - событие выбор 4 мячей из ящиков, 2 - событие 2 мяча из 4-x выбранных - белые
# Вероятность попадания какого либо мяча из 1 или 2 ящика составляет 1/2 так как берут 2 мячя из одного и 2 из другого
# Вероятность что взятый из первого ящика мяч будет белый составляет 7/10, из второго 9/11
# По формуле полной вероятности, вероятность вытащить один белый мяч будет P = 1/2*7/10+1/2*9/11
# Определив вероятность, можно применить формулу Бернулли, чтобы определить вероятность для заданного количества мячей в серии из 4 испытаний
pW = (1/2*box1['w']/bs1)+ (1/2*box2['w']/bs2)
print("Вероятность, что 1  попавший в выборку мяч белый ",pW)

pW2from4 = p_binomial(2, # в выборке из 4 мячей оказолось 2 белых
                      4, # максимум может выпасть 4 мяча 
                      pW) # Вероятность,  что взятый мяч будет белый
print("Ответ: Вероятность, что ровно 2 из 4 мячей белые", pW2from4)
```

```
# Аналогично определим вероятность того, что хотя-бы один мяч белый
# Условие хотя бы один мяч предполагает вероятность неблагоприятного исхода, только для случая если в выборке из 4 мячей будет 0 белых
# В остальных случаях в выборке будут белые мячи
# Определим вероятность отсутствия мячей в выборке

pW0from4 = p_binomial(0, # случайная величина примет значение 0 из 4 испытаний
                      4, pW)
print("Вероятность неблагоприятного исхода - 0 белых мячей", pW0from4)    

#Сумма вероятностей всхех возможных исходов равна 1 
print("Ответ: Вероятность, что хотябы 1 мяч из 4 белый составляет", 1-pW0from4)
```

```
# Аналогично можно обобщить решение и для первого случая - 4 из 4 взятых мячей белые

pW4from4 = p_binomial(4,4,pW)
print("Ответ: Вероятность что все мячи из 4 белые", pW4from4)

print("Проверка первого решения разность должна -> к нулю d=", pW4from4-pAB1)
```
