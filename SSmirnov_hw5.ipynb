```
import numpy as np
from scipy.stats import norm, t # функции распределений нормальное и стьюдента
```

<h3> Задача1</h3>
<p> Известно, что генеральная совокупность распределена нормально со средним квадратическим отклонением, равным 16. Найти доверительный интервал для оценки математического ожидания a с надежностью 0.95, если выборочная средняя M = 80, а объем выборки n = 256.</p>

```
# Стандартное отклонение генеральной совокупности известно, можно применить z распределение для расчета интервала 
std1 = 16
m1 = 80
n1 = 256

z_05a = 1-0.05/2 # значение z - функции
print('Расчетное значение z- функции',z_05a)

Z1 = norm.ppf(z_05a) # подходящее значение x из z - функции
print('Z1 ', round(Z1,2))
# границы интервала
t1 = Z1 * (std1/n1**0.5)
print('Границы интервала', round(t1,4))
print(f"Ответ истиное математическое ожидание с 95% вероястностью находится в интервале от {round(m1-t1,4)} до {round(m1+t1,4)}")
```

<h3> Задача2</h3>
<p> В результате 10 независимых измерений некоторой величины X, выполненных с одинаковой точностью, получены опытные данные: 6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1 Предполагая, что результаты измерений подчинены нормальному закону распределения вероятностей, оценить истинное значение величины X при помощи доверительного интервала, покрывающего это значение с доверительной вероятностью 0,95.</p>

```
x2 = np.array([6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1], dtype = float)
print(x2)
print("размерность массива", x2.shape)
```

```
# для решения задачи будем использовать критерий стьюдента
m2 = x2.mean() # оценка среднего выборочного значения
std2 = x2.std(ddof=1) # неспещенная оценка среднеквадратичного отклонения
n2 = int(x2.shape[0])

print(f"Описательная статистика для X: \n Среднее выборочное: {m2} \n сигма: {std2} \n размер выборки: {n2} \n параметр дов. интервала: {z_05a}")
```

```
T2 = t.ppf(z_05a, 
            n2-1 ) # оценка степеней свободы

print("Значение оргумента распределения стьюдента", round(T2,4))
```

```
# расчитаем границы доверительного интервала

t2 = T2 * std2/n2**0.5
print(f"Ответ: Истинное значение величины Х c вероятностью 95% находится в интервале от {round(m2-t2,4)} до {round(m2+t2,4)}" )
```

<h3> Задача3</h3>
<p> Утверждается, что шарики для подшипников, изготовленные автоматическим станком, имеют средний диаметр 17 мм.
Используя односторонний критерий с α=0,05, проверить эту гипотезу, если в выборке из n=100 шариков средний диаметр
оказался равным 17.5 мм, а дисперсия известна и равна 4 кв.мм.<p>

```
# имеется дисперсия генеральной совокупности, решаем и использование z - распределения
# H0 - шарики имеют средний диаметр 17 мм, H1 - шарики имеют диаметр d > 17мм
# поддтвердим истинность нулевой гиподезы на уровне значимости 5%

#расчитаем наблюдаемый критерий
Zn3 = (17.5-17.0)/(4**0.5/100**0.5)
print("Значение Z наблюдаемого", Zn3)

Zk3 = norm.ppf(1-0.05)
print("Значение Z критического", Zk3)

print("Ответ: Нулевая гипотеза не нашла подтверждение так как критический критерий оказался меньше наблюдаемого ")
```

```
#проверка - вероятность что в генеральной совкупность со средним 17мм и ст. отконением 2 втретить шарики диаметром больше 17,4

print(1 - norm.cdf(17.4, loc = 17.0, scale = 2))
```

```
#Вероятность встретить данные шарики в генеральной совокупности со средним 17.2 и ст. отклонением 2
print(1 - norm.cdf(17.4, loc = 17.2, scale = 2))

# Раз вероятность встретить данные шарики в распределении с большим средним значением выше, то альтернативная гипотеза косвенно подтверждаестся
# мы смещаемся в сторону распределения новой генеральной совокумности, в противном случае было бы наоборот
print(1 - norm.cdf(17.4, loc = 16.9, scale = 2))
```

<h3> Задача4</h3>
<p> Продавец утверждает, что средний вес пачки печенья составляет 200 г.
Из партии извлечена выборка из 10 пачек. Вес каждой пачки составляет:
202, 203, 199, 197, 195, 201, 200, 204, 194, 190.
Известно, что их веса распределены нормально.
Верно ли утверждение продавца, если учитывать, что доверительная вероятность равна 99%?</p>

```
# Дисперсия генеральной совокумности не дана, для решения задачи применим распределение стьюдента
x4 = np.array([202, 203, 199, 197, 195, 201, 200, 204, 194, 190])
m4 = x4.mean() # выборочное среднее значение 
std4 = x4.std(ddof = 1)
n4 = len(x4)
a4 = (1-0.99)/2
print(f"""Описательная статистика для набора пачек: \n 
Среднее выборочное: {m4} \n
 сигма: {std4} 
\n размер выборки: {n4} \n
параметр дов. интервала: {a4}""")
```

```
# расчитаем доверительный интервал
T4 = t.ppf(1-a4,  # квантиль t распределения
             n4-1) # оценка степени свободы
print("Значение критерия t распределения", round(T4,4))

# оценка интервала
t4 = T4 * std4/n4**0.5
print("Границы интервала +-", round(t4))

print(f"Среднее значение генеральной совокупности с вероятностью 99% находится в интервале от {round(m4-t4,4)} от {round(m4+t4,4)}")
print("Ответ: утверждение продавца верно с вероятностью 99% посколько значение веса пачки в 200 г находится внутри доверительного интервала")
```
