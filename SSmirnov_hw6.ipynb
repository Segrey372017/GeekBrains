```
import numpy as np
import pandas as pd
from scipy.stats import norm, t
```

<h3> Задача1 </h3>
<p> Даны значения величины заработной платы заемщиков банка (zp) и значения их поведенческого кредитного скоринга (ks):
zp = [35, 45, 190, 200, 40, 70, 54, 150, 120, 110],
ks = [401, 574, 874, 919, 459, 739, 653, 902, 746, 832].
<ul> <li> Найдите ковариацию этих двух величин с помощью элементарных действий, а затем с помощью функции cov из numpy
Полученные значения должны быть равны. </li>
<li> Найдите коэффициент корреляции Пирсона с помощью ковариации и среднеквадратичных отклонений двух признаков,
а затем с использованием функций из библиотек numpy и pandas. </li> </ul></p>

```
zp = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110]) # инициализируем массивы
ks = np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])
print(f" zp = {zp}, \n ks = {ks}")
```

```
cov1 = np.mean(zp*ks) - np.mean(zp)*np.mean(ks) # вычисляем ковариацию по формуле
print("Корреляция с помощью элементарных действий", cov1)
```

```
cov11 = np.cov(zp, ks, ddof=0)
print("Ковариация с помощью библиотечной функции", cov11[0,1])
```

```
# Коэффициент корреляции Пирсона
std_zp = np.std(zp)
std_ks = np.std(ks)

kcorr1 = cov1/(std_zp*std_ks)
print("Коэффициент корреляции с помощью элементарных действий", kcorr1)
```

```
kcorr11 = np.corrcoef(zp, ks)
print("Коэффициент корреляции по библиотечной функции", kcorr11[0,1])
```

<h3> Задача2</h3>
<p> Измерены значения IQ выборки студентов,
обучающихся в местных технических вузах:
131, 125, 115, 122, 131, 115, 107, 99, 125, 111.
Известно, что в генеральной совокупности IQ распределен нормально.
Найдите доверительный интервал для математического ожидания с надежностью 0.95.</p>

```
#Дисперсия генеральной совокупности неизвестна используем распределение стьюдента
st = np.array([131, 125, 115, 122, 131, 115, 107, 99, 125, 111])
m2 = st.mean() # среднее выборочное IQ студентов
std2 = st.std(ddof=1) # несмещенное стандартное отклонение выборки
t_95 = 1 - (1-0.95)/2  # значения t распределени для выбранного интервала
n2 = len(st)
print(f""" Описательная статистика для выборки студентов:
       Среднее значение: {m2}
       Стандартное отклонение: {std2} 
       Число наблюдений в выборке: {n2}
       Квантиль t распределения {t_95}""")
```

```
#Оценка доверительного интервала
t2 = t.ppf(t_95, # Квантиль распределения
           n2-1) # степень свободы выборки
print("Значение аргумента t-распределенияя ", round(t2,4))

t_iq = t2*std2/np.sqrt(n2)
print("Граница интервала ", t_iq)


print(f"Математическое ожидание генеральной совокупности с вероятностью 95% находится в диапазоне от {round(m2-t_iq,4)} до {round(m2+t_iq,4)}")
```

<h3>  Задача3</h3>
<p> Известно, что рост футболистов в сборной распределен нормально
с дисперсией генеральной совокупности, равной 25 кв.см. Объем выборки равен 27,
среднее выборочное составляет 174.2. Найдите доверительный интервал для математического
ожидания с надежностью 0.95.<p3>

```
# Дисперсия генеральной совокупности известана, для решения используем z - распределение
std3 = 25**0.5 # среднее квадратическое отклонение 
n3 = 27 # объем выборки
m3 = 174.2  # среднее выборочное
z_95 = t_95 # квантиль z распределения

z3 = norm.ppf(z_95)
print("Аргумент z распределения", z3)

z_ft = z3 * std3/n3**0.5
print("Граница интервала ", z_ft)
print(f"Доверительный интервал с надежностью 95% находится в границах от {round(m3-z_ft,4)} до {round(m3+z_ft,4)}")
```
